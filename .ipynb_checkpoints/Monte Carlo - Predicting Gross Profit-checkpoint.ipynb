```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rev_m = 170 #millions of dollars
rev_stdev = 20
iterations = 1000
```

```python
rev = np.random.normal(rev_m, rev_stdev, iterations) #normal distribution
rev
```

```python
plt.figure(figsize=(15,6))
plt.plot(rev)
plt.show()
```

```python
COGS = - (rev * np.random.normal(0.6,0.1))# normally Cost of Goods sold should be ~60% of the revenues, money spent (-)

plt.figure(figsize=(15,6))
plt.plot(COGS)
plt.show()
```

```python
 COGS.mean()
```

```python
COGS.std() 
```

```python
Gross_Profit = rev + COGS
Gross_Profit

plt.figure(figsize=(15,6))
plt.plot(Gross_Profit)
plt.show()
```
